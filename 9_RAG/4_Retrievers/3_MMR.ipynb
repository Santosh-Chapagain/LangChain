```python
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embedding = HuggingFaceEmbeddings(
    model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```python
documents = [

    Document(
        page_content="Python is widely used for data analysis and machine learning projects.",
        metadata={"id": "doc1"}
    ),

    Document(
        page_content="Python is popular in data science because of libraries like pandas and NumPy.",
        metadata={"id": "doc2"}
    ),

    Document(
        page_content="Python is used to build machine learning models and AI applications.",
        metadata={"id": "doc3"}
    ),

    Document(
        page_content="Java is commonly used for enterprise software and backend systems.",
        metadata={"id": "doc4"}
    ),

    Document(
        page_content="Data scientists use Python to clean data and train prediction models.",
        metadata={"id": "doc5"}
    ),

    Document(
        page_content="Java is also used to build large-scale applications in companies.",
        metadata={"id": "doc6"}
    ),

]
```

```python
vector_stores =  FAISS.from_documents(
    embedding=embedding,
    documents= documents
)
```

```python
retriever = vector_stores.as_retriever(
    search_type = 'mmr', # this enable mmr
    search_kwargs={'k': 3, "lambda_mult": 0.5}  # lambda_mult , value lies between (0-1). If set 1 then , it become similar to similarity search but when set 0 or near 0 value then it give diverse result .
)
```

```python
query = "Where python is mostly used ?"
result = retriever.invoke(query)
```

```python
for i , doc in enumerate(result):
    print(f"... Result {i+1} ...\n")
    print(f"Ans=> {doc.page_content}\n")
```
